```python
for i in range(50):
    dir = "segment" + str(i).zfill(2)
    for label in [os.path.join(dir + "/labels", x) for x in sorted(os.listdir(dir + "/labels"))]:
        lines = open(label).readlines()
        with open(label, "w") as f:
            for line in lines:
                if float(line.split()[1]) < 0:                          # if left is negative, make it 0
                    line = line.replace(line.split()[1], "0.000000")
                if line[0] == '2':
                    line = '0' + line[1:]
                elif line[0] == '4':
                    line = '1' + line[1:]
                elif line[0] == '1':
                    line = '2' + line[1:]
                else:
                    print("SIGN")

                f.write(line)
```

```python
for i in range(50):
    dir = "segment" + str(i).zfill(2)
    for label in [os.path.join(dir + "/labels", x) for x in sorted(os.listdir(dir + "/labels"))]:
        lines = open(label).readlines()
        with open(label.replace("labels", "yolo_labels"), "w") as f:
            for line in lines:
                left = float(line.split()[1])
                top = float(line.split()[2])
                right = float(line.split()[3])
                bottom = float(line.split()[4])

                centre_x = "{:.6f}".format((right + left) / 2)
                centre_y = "{:.6f}".format((top + bottom) / 2)
                width = "{:.6f}".format(right - left)
                height = "{:.6f}".format(top - bottom)

                line = line[0]+"  "+centre_x+"  "+centre_y+"  "+width+"  "+height+"\n"

                f.write(line)
```

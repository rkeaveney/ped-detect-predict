```python
for i in range(0, 5):
    n = str(i)
    image = cv2.imread('segment11/images/s11_f000'+n+'.jpg')
    with open('segment11/labels/s11_f000'+n+'.txt') as f:
        lines = f.readlines()
        for line in lines:
            if line.split()[0] == "0":
                colour = (0,0,255)
            elif line.split()[0] == "1":
                colour = (255,0,0)
            elif line.split()[0] == "2":
                colour = (0,255,0)
            x1 = int(floor(float(line.split()[1]) * 1920))
            y1 = int(floor(float(line.split()[2]) * 1920))
            x2 = int(floor(float(line.split()[3]) * 1920))
            y2 = int(floor(float(line.split()[4]) * 1920))
            cv2.rectangle(image, (x1, y1), (x2,y2), colour, 4)
    cv2.imwrite('s00_f000'+n+'_bb.jpg', image)
    files.download('s00_f000'+n+'_bb.jpg') 
```
